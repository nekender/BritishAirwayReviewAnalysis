```python
for url in links:
    driver.get(url)
    content=driver.page_source
    soup = BeautifulSoup(content)
    for article in soup.findAll(attrs={'itemprop':'review'}):
        head=article.find('h2',attrs={'class':'text_header'})
        review=article.find('div',attrs={'class':'text_content'})
        rating=article.find('span',attrs={'itemprop':"ratingValue"})
        reviews.append(review.text)
        ratings.append(rating)
        headers.append(head.text)
```

```python
data=pd.DataFrame({'Title':headers,'Review':reviews,'Rating':ratings})
```

# DATA PREPROCESSING


```python
for i in range(3399):
    data['Review']=data['Review'].apply(lambda x: x.lower())
```

```python
import re
for i in range(3399):
    data['Review']=data['Review'].apply(lambda x: re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?✅", "", x))
```

```python
data
```

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = stopwords.words('english')

for i in range(3399):
    data['Review']=data['Review'].apply(lambda x:" ".join(word for word in x.split() if word not in (stop)))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

stemmer=PorterStemmer()
for i in range(3399):
    data['Review']=data['Review'].apply(lambda x:stemmer.stem(x))
```

```python
import nltk 

for i in range(3399):
    data['tokens']=data['Review'].apply(lambda x: nltk.word_tokenize(x))
```
